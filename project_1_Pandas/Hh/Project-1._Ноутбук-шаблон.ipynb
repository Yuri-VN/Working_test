# Преобразование данных

1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# Создаем копию базы и аналогично далее по мере преобзования данных
data_transformation_1 = hh_data.copy()

# функция разбивает значение признака на элементы, затем берем первые 2 элемента и объединяем их. Далее возвращаем необходимые значения в новый признак с учетом требуемых 4-х категорий.
def get_education(z):
    get_ed = ' '.join(z.split()[:2])
    if 'Высшее' in get_ed:
        return 'высшее'
    if 'Неоконченное высшее' in get_ed:
        return 'неоконченное высшее'
    if 'Среднее специальное' in get_ed:
        return 'среднее специальное'
    if 'Среднее' in get_ed:
        return 'среднее'
# создаем новый признак 'Образование'применяя метод apply к столбцу 'Образование и ВУЗ'
data_transformation_1['Образование'] = data_transformation_1['Образование и ВУЗ'].apply(get_education)

# Проверяем правильность созданого признака - 'Образование'
display (data_transformation_1['Образование'].unique())

# Удаляем признак 'Образование и ВУЗ'
data_transformation_1 = data_transformation_1.drop('Образование и ВУЗ', axis = 1)

# Отвечаю на контрольные вопросы в задании. Поскольку эта информация далее не будет применена код закомментировал
#display (data_transformation_1['Образование'].value_counts()['среднее'])        
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
data_transformation_2 = data_transformation_1.copy()

# функция разбивает значение признака на элементы, затем берем [0] элемент. Далее возвращаем новые уникальные значения в новый признак - 'Пол'
def get_gender(z):
    get_gen = z.split(',')[0].lower() # .lower() - здесь ни на что не влияет, ради тренировки
    if 'мужчина' in get_gen:
        return 'М'
    if 'женщина' in get_gen: # можно просто else, выбрал If т.к. мне так нагляднее
        return 'Ж' 
# создаем новый признак 'Пол'   
data_transformation_2['Пол'] = data_transformation_2['Пол, возраст'].apply(get_gender)

# Проверяем правильность созданого признака 'Пол'
display (data_transformation_2['Пол'].unique())


# функция разбивает значение признака на элементы, затем берем сначала [1] элемент, далее этот элетент сплитуем и берем [0] элемент. Возвращаем последний элемент преобразовав его в int
def get_age(z):
    get_ag = z.split(',')[1].split()[0]
    return int(get_ag)
# создаем новый признак 'Возраст'
data_transformation_2['Возраст'] = data_transformation_2['Пол, возраст'].apply(get_age)

# Удаляем признак 'Пол, возраст'
data_transformation_2 = data_transformation_2.drop('Пол, возраст', axis = 1)

# Отвечаю на контрольные вопросы в задании. Поскольку эта информация далее не будет применена код закомментировал
#display (round(data_transformation_2['Пол'].value_counts(normalize=True)*100,2))
#display (round(data_transformation_2['Возраст'].mean(),1))
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
data_transformation_3 = data_transformation_2.copy()

# функция сначала берет пропуски и скрытые пропуски 'Не указано', после чего возвращает None
def get_experience(z):
    if z is np.nan or z == 'Не указано':
        return None
    
    # затем функция разбивает значение признака на элементы, затем берем первые 5 элементов
    get_experience_split = z.split(' ')[:6]
    # формируем списки с годами и месяцами
    year_list = ['год', 'года', 'лет']
    month_list = ['месяц', 'месяца', 'месяцев']
    year = 0
    month = 0
    # Используя цикл и зная, что перед текстом всегда идет число, применяем функцию enumerate. Далее отнимаем 1 от индекса и возвращаем год, месяц не забыв год умножить на 12 месяцев 
    for index, item in enumerate (get_experience_split):
        if item in year_list:
            year = int(get_experience_split[index-1])
        if item in month_list:
            month = int(get_experience_split[index-1])
    return year*12 + month    
# создаем новый признак 'Опыт работы (месяц)'
data_transformation_3['Опыт работы (месяц)'] = data_transformation_3['Опыт работы'].apply(get_experience)

# Удаляем признак 'Опыт работы'
data_transformation_3 = data_transformation_3.drop('Опыт работы', axis = 1)

# Отвечаю на контрольные вопросы в задании. Поскольку эта информация далее не будет применена код закомментировал
#display (round (data_transformation_3['Опыт работы (месяц)'].median()))
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
data_transformation_6 = data_transformation_5.copy()

# Читаем данные, создаем копию базы, удаляем ненужные приизнаки
hh_exchangeRates = pd.read_csv('../data/hh_data/ExchangeRates/ExchangeRates.csv', sep=',')
hh_exchangeRates_1 = hh_exchangeRates.copy()
hh_exchangeRates_1 = hh_exchangeRates_1.drop(columns=['per', 'time', 'vol'], axis = 1)

# Преобразуем признаки 'Обновление резюме' и 'date' в формат datetime. Укажем параметр dayfirst=True поскольку в первоначальном признаке первым идет день. Применим атрибут date
data_transformation_6['Обновление резюме'] = pd.to_datetime(data_transformation_6['Обновление резюме'], dayfirst=True).dt.date
hh_exchangeRates['date'] = pd.to_datetime(hh_exchangeRates['date'], dayfirst=True).dt.date


# Функция разбивает значение признака на элементы и берет [0] элемент. Далее этот элемент преобразуем в float и возвращаем в новый признак - 'Сумма ЗП'
def get_salary(z):
    salary =  z.split(' ')[0] 
    return float(salary)
# создаем новый признак 'Сумма ЗП'
data_transformation_6['Сумма ЗП'] = data_transformation_6['ЗП'].apply(get_salary)


# Функция разбивает значение признака на элементы и берет [1] элемент. Убираем точки в значение [1] элемента. С помощью условного оператора возвращаем преобразованное значение в новый признак 'Валюта ЗП'
def get_currency(z): 
    # используя словарь сокращаем код, т.е. не используем большое количество if
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
        }
    currency =  z.split(' ')[1].replace('.', '') # можно из без  - replace - , только тогда необходимо оставить точки как они есть (руб., грн., бел.руб.)
    if currency == 'руб':
        return 'RUB'
    else:
        return currency_dict[currency] 
# создаем новый признак 'Валюта ЗП'
data_transformation_6['Валюта ЗП'] = data_transformation_6['ЗП'].apply(get_currency)

# Проверяем правильность созданого признака 'Валюта ЗП'
display(data_transformation_6['Валюта ЗП'].unique())

# Объединяем таблицу с резюме и с курсами по столбцам с датой и названием валюты. Тип объединения применяем - left.
merged_data = data_transformation_6.merge(
    hh_exchangeRates_1,
    left_on=['Обновление резюме', 'Валюта ЗП'],
    right_on=['date', 'currency'],
    how='left')

# Заполняем значения единицей 1 для признаков 'close' и 'proportion' применяя метод fillna
merged_data['close'] = merged_data['close'].fillna(1)
merged_data['proportion'] = merged_data['proportion'].fillna(1)

# Производим математические операции со столцами и создаем новый столбец/признак 'ЗП (руб)'
merged_data['ЗП (руб)'] = merged_data['Сумма ЗП'] * merged_data['close'] / merged_data['proportion']

# Отвечаю на контрольные вопросы в задании. Поскольку эта информация далее не будет применена код закомментировал
# есть отклонение на 1 тыс.руб. : эталон 59 у меня 60
# display (round(merged_data['ЗП (руб)'].median()/1000,))

# Удаляем признаки 'ЗП', 'Сумма ЗП', 'Валюта ЗП', 'proportion', 'currency', 'date', 'close'
merged_data = merged_data.drop(['ЗП', 'Сумма ЗП', 'Валюта ЗП', 'proportion', 'currency', 'date', 'close'], axis = 1)

# Выводим данные для сверки с эталонным решением
display (merged_data.info())
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
fig = plt.subplots(figsize=(8, 4))

log_age = np.log(drop_data_transformation_7['Возраст'])

histplot = sns.histplot(
    log_age,
    bins=30,
    )
histplot.axvline(log_age.mean(), color='r', lw=2)
histplot.axvline(log_age.mean()+ 3 * log_age.std(), color='r', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='r', ls='--', lw=2)
histplot.set_title('Log Возраст');

print(round(log_age.skew(),2))


def outliers_z_score(data, sign, log_scale=False):
    if log_scale:
        x = np.log(drop_data_transformation_7[sign])
    else:
        x = drop_data_transformation_7[sign]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers_3 = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers_3, cleaned

outliers_3, cleaned = outliers_z_score(drop_data_transformation_7, 'Возраст', log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers_3.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')

display (outliers_3['Возраст'])

drop_data_transformation_7 = drop_data_transformation_7.drop(outliers_3.index, axis=0)
```
